```python
from vision_studio.data_loader import SimpleDataLoader
from vision_studio.dataset import MnistDataset
from vision_studio.models import BaseModel

import torch
import torch.nn as nn
from torch import Tensor
from typing import Any
```

```python
dataset = MnistDataset("/home/local/.cache/kagglehub/datasets/hojjatk/mnist-dataset/versions/1")
```

```python
dataloader = SimpleDataLoader(dataset, batch_size=32, shuffle=True)
```

```python
class CNNModel1(BaseModel):
    def __init__(self, in_channels: int, num_classes: int):
        super().__init__()
        self.in_channels = in_channels
        self.num_classes = num_classes
        self.features = nn.Sequential(
            nn.Sequential(
            nn.Conv2d(in_channels, 16, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
            ),
            nn.Linear(16, 64),
            nn.ReLU(),
            nn.Linear(64, num_classes),
        )
    
    def forward(
        self,
        input: Tensor,
        targets: dict[str, Any] | None = None,
    ):
        logits = self.features(input)
        probs = torch.softmax(logits, dim=1)
        labels = torch.argmax(probs, dim=1)
        
        return {
            "logits": logits,
            "probs": probs,
            "labels": labels,
        }
    
    def compute_loss(
        self,
        outputs: dict[str, Any],
        targets: dict[str, Any],
    ) -> Tensor:
        logits = outputs["logits"]
        labels = targets["labels"]
        loss_fn = nn.CrossEntropyLoss()
        loss = loss_fn(logits, labels)
        return loss
        
    def get_config(self) -> dict[str, Any]:
        return {
            "in_channels": self.in_channels,
            "num_classes": self.num_classes,
        }

class CNNModel2(BaseModel):
    def __init__(self, in_channels: int, num_classes: int):
        super().__init__()
        self.in_channels = in_channels
        self.num_classes = num_classes
        self.features = nn.Sequential(
            nn.Sequential(
            nn.Conv2d(in_channels, 16, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
            ),
            nn.Linear(16, num_classes),
        )
    
    def forward(
        self,
        input: Tensor,
        targets: dict[str, Any] | None = None,
    ):
        logits = self.features(input)
        probs = torch.softmax(logits, dim=1)
        labels = torch.argmax(probs, dim=1)
        
        return {
            "logits": logits,
            "probs": probs,
            "labels": labels,
        }
    
    def compute_loss(
        self,
        outputs: dict[str, Any],
        targets: dict[str, Any],
    ) -> Tensor:
        logits = outputs["logits"]
        labels = targets["labels"]
        loss_fn = nn.CrossEntropyLoss()
        loss = loss_fn(logits, labels)
        return loss
        
    def get_config(self) -> dict[str, Any]:
        return {
            "in_channels": self.in_channels,
            "num_classes": self.num_classes,
        }
    
class CNNModel3(BaseModel):
    def __init__(self, in_channels: int, num_classes: int):
        super().__init__()
        self.in_channels = in_channels
        self.num_classes = num_classes
        self.features = nn.Sequential(
            nn.Sequential(
            nn.Conv2d(in_channels, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
            ),
            nn.Linear(16, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.Linear(32, num_classes),
        )
    
    def forward(
        self,
        input: Tensor,
        targets: dict[str, Any] | None = None,
    ):
        logits = self.features(input)
        probs = torch.softmax(logits, dim=1)
        labels = torch.argmax(probs, dim=1)
        
        return {
            "logits": logits,
            "probs": probs,
            "labels": labels,
        }
    
    def compute_loss(
        self,
        outputs: dict[str, Any],
        targets: dict[str, Any],
    ) -> Tensor:
        logits = outputs["logits"]
        labels = targets["labels"]
        loss_fn = nn.CrossEntropyLoss()
        loss = loss_fn(logits, labels)
        return loss
        
    def get_config(self) -> dict[str, Any]:
        return {
            "in_channels": self.in_channels,
            "num_classes": self.num_classes,
        }
        
        
class CNNModel4(BaseModel):
    def __init__(self, in_channels: int, num_classes: int):
        super().__init__()
        self.in_channels = in_channels
        self.num_classes = num_classes
        self.features = nn.Sequential(
            nn.Sequential(
            nn.Conv2d(in_channels, 16, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout(2)
            ),
            nn.Linear(16, 64),
            nn.Linear(64, num_classes),
        )
    
    def forward(
        self,
        input: Tensor,
        targets: dict[str, Any] | None = None,
    ):
        logits = self.features(input)
        probs = torch.softmax(logits, dim=1)
        labels = torch.argmax(probs, dim=1)
        
        return {
            "logits": logits,
            "probs": probs,
            "labels": labels,
        }
    
    def compute_loss(
        self,
        outputs: dict[str, Any],
        targets: dict[str, Any],
    ) -> Tensor:
        logits = outputs["logits"]
        labels = targets["labels"]
        loss_fn = nn.CrossEntropyLoss()
        loss = loss_fn(logits, labels)
        return loss
        
    def get_config(self) -> dict[str, Any]:
        return {
            "in_channels": self.in_channels,
            "num_classes": self.num_classes,
        }           
        
            
        
        
```
